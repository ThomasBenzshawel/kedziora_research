```python
import pandas as pd
import seaborn as sns
```

```python
df_thingiverse = pd.read_csv('thingiverse_stls.csv')
df_modelnet = pd.read_csv('modelnet_stls.csv')
df_corrupt = pd.read_csv('corrupt_thingiverse_stls.csv')
```

```python
df_thingiverse['passed'] = df_thingiverse['score'].apply(lambda v: v == 15)
df_modelnet['passed'] = df_modelnet['score'].apply(lambda v: v == 15)
df_corrupt['passed'] = df_corrupt['score'].apply(lambda v: v == 15)
```

```python
df_modelnet['no_change_manifold'] = (df_modelnet['manifold'] == True) & (df_modelnet['score'] == 15)
df_corrupt['no_change_manifold'] = (df_corrupt['manifold'] == True) & (df_corrupt['score'] == 15)
```
